Note: It is possible that better hyperparameters exist for certain environments. Results are also very sensitive to the seed used. The "Humanoid" tasks/environments were omitted since they required considerably more enivronment steps to converge (~20M steps vs. 1M steps for most environments) and it would multiple runs to find good hyperparameters for them.

```python
import warnings
warnings.filterwarnings('ignore')

from src.agent import PPO_Agent
from baselines.common import plot_util as pu
```

### Setting PPO Hyperparameters and seed for environments & libraries

```python
model_dir = "models"
log_dir = "logs"
plot_dir = "plots"

params = {
    'NUM_ENV_TIMESTEPS' : 1e6,
    'NUM_TIMESTEPS_PER_UPDATE' : 2048,
    'NUM_MINIBATCHES' : 32,
    'NUM_EPOCHS' : 4,

    'GAMMA' : 0.99,
    'CLIP_PARAM' : 0.2,
    'LAMBDA' : 0.95,
    'VALUE_FUNCTION_COEF' : 0.5,
    'ENTROPY_COEF' : 0,

    'ACTOR_HIDDEN_UNITS' : [64, 64], 
    'CRITIC_HIDDEN_UNITS' : [64, 64],
    'LEARNING_RATE' : 3e-4,
    'OPTIMIZER_EPSILON' : 1e-7,

    'SAVE_INTERVAL' : 25
}
```
